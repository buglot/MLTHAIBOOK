จะใช้ model หsoftmax regression

```python
def softmax(z):
    z = z - np.max(z, axis=1, keepdims=True)  # ลบ max แต่ละแถว
    e = np.exp(z)
    return e / np.sum(e, axis=1, keepdims=True)
```

```python
def compute_gradient(X, y_true, y_hat):
    m = X.shape[0]
    dW = (1/m) * X.T @ (y_hat - y_true) 
    return dW.T
```

```python
def cross_entropy_loss(y_true, y_pred):
    eps = 1e-15
    y_pred_clipped = np.clip(y_pred, eps, 1 - eps)

    m = y_true.shape[0]
    loss = -np.sum(y_true * np.log(y_pred_clipped)) / m

    return loss
```

```python
y = data_iris.target
y
```

```python
y_one_hot = np.zeros((y.size, y.max()+1 ))
y_one_hot[np.arange(y.size), y] = 1
y_one_hot 
```

```python
learning_rate = 1e-2
weights=np.ones((y_one_hot.shape[1],data_iris.data.shape[1])) 
weights
```

```python
weights=np.ones((y_one_hot.shape[1],data_iris.data.shape[1])) 
for epoch in range(9000):
    z = data_iris.data @ weights.T
    y_hat = softmax(z)
    loss = cross_entropy_loss(y_one_hot, y_hat)
    
    dw = compute_gradient(data_iris.data,y_one_hot, y_hat)
    weights -= learning_rate * dw
    if epoch%10 ==0 :
        print(f"{epoch}/4000 loss:",loss)
```

```python
def predict(data,weights):
    z = data @ weights.T  # คำนวณ logits (shape = (4,))
    y_hat = softmax(z)          # probability ของแต่ละคลาส
    predicted_class_index = np.argmax(y_hat)
    predicted_label = data_iris.target_names[predicted_class_index]
    print("Predicted label:", predicted_label)
```

# หาค่าความแม่นยำ accuracy

$$accuracy = \frac{TP+TN}{TP+TN+FP+FN}$$

โดยเราจะมองว่า TP เป็นอะไร ที่เหลือเป็น TN

อธิบาย

TP = True Positive คือหลักการง่าย aiของเราเดา`ถูกต้อง` แล้วเดาเป็น Positive

TN = True Negative คือ aiของเราเดา`ถูกต้อง` แล้วเดาเป็น Negative

FP = False Positive คือ aiของเราเดา`ผิด` แล้วเดาเป็น Positive

FN = False Negative คือ aiของเราเดา`ผิด` แล้วเดาเป็น Negative

หรือการคำนวนง่ายๆของ accuracy คือ อันไหนเดาถูกต้อง มาการจำนวนทั้งหมด และ`ใช้สำหรับ class มากกว่า 2 ด้วย`

$$accuracy = \frac{เดาถูกต้อง}{จำนวนทั้งหมด}$$

ก่อนหน้าผมให้มอง TP เป็นอะไรแล้วที่ที่เหลือเป็น TN คือ 

จริงแล้วจะให้มอง P และ N

# Precision คืออะไร

สมการ Precision 

$$
\frac{TP}{TP+FP}
$$

หมายความว่ายังไง

TP คือ aiของเราเดา`ถูกต้อง` เดาเป็น Positive 

FP คือ aiของเราเดา`ผิด` เดาเป็น Positive 

จากสมการมองให้เห็นว่า Precision คือ aiเดาถูกต้องที่เป็น p หารด้วย aiเดาถูกต้องที่เป็น p + ai เดาผิดที่เป็น p

หรือ ai เดาถูกเป็น p/ค่าที่ ai เดาเป็น p ผิดทั้งหมด

ชึ่งจากสมการ ผมมองว่า Precision คือ  `โอกาสที่ ai เดาเป็น P ถูกต้องกี่เปอร์เซ็น`

# recall คืออะไร

สมการ คือ 

$$
\frac{TP}{TP+FN}
$$

TN = หมายความว่า ai เดาถูก ที่เป็น N เสมอไป โดยแนะนำให้มองเป็นแล้วอยากนี้เมื่อ class มากกว่า 2

โดยจากรู้และสมการ ค่าเดาถูกที่เป็น P/ค่าความถูกต้องทั้งหมดของ P

การสมการแล้วผมว่า recall คือ จาก class P ทั้งหมด ai เดาถูกกี่เปอร์เซ็น
