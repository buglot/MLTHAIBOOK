# gradient descent and stochastic gradient descent

ต่างกันยังไง


| ส่วน                       | GD                       | SGD                    |
| -------------------------- | ------------------------ | ---------------------- |
| การคำนวณ gradient          | ใช้ข้อมูลทั้งหมดพร้อมกัน | ใช้ทีละตัวอย่าง        |
| การอัปเดตพารามิเตอร์       | อัปเดต 1 ครั้งต่อ epoch  | อัปเดตทุกตัวอย่าง      |
| ความนิ่งของ loss           | สูงกว่า                  | มี noise, กระโดดขึ้นลง |
| ความเร็ว (เมื่อข้อมูลเยอะ) | ช้า                      | เร็วกว่า               |


ตัวอย่างโค้ดแบบ numpy gradient descent ใน linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
X = np.random.rand(100, 1)  # (100, 1)
y = 3 * X + 2 + 0.1 * np.random.randn(100, 1)
w = np.ones(1)
b = np.ones(1)
lr = 0.1  
epochs = 100
```

# data

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
```

# train GD

```python
for i in range(epochs):
    y_hat =w*X+b
    error = y_hat-y
    dw = np.mean(2*X*error)
    db = np.mean(2 * error)
    w -= lr * dw
    b -= lr * db
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,w*X+b,color="red",linewidth=3)
```

# train  SGD

SGD การอัปเดตค่า ทีละดาต้าที่สุ่มมา

```python
w = np.ones(1)
b = np.ones(1)
```

```python
for i in range(epochs):
    indices = np.random.permutation(len(X))
    for j in indices:
        xi = X[i]
        yi = y[i]
        y_hat =w * xi + b
        error = y_hat-yi
        dw = 2*xi*error
        db = 2 * error
        w -= lr * dw
        b -= lr * db
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X,y)
plt.plot(X,w*X+b,color="red",linewidth=3)
```
